# Histogram Oriented Gradient with Dlib
This code show how to detect faces with some functions of Dlib package. In this package we can find same high level function like get_frontal_face_detector().
This function is a Histogram Oriented Gradient implementation. The HOG is a feature descriptor that works with the image texture gradient. This gradient results in some like the edges of the image. 
It is interesting to inform that the coordinates data of bounding boxes in Dlib are different of OpenCV. In Dlib the bounding boxes are represented be coordinates of two point that trace a diagonal of the bounding box.
Below is shown a simple example of using: 

[Dlib Documentation](http://dlib.net/python/index.html#dlib.get_frontal_face_detector)

[More about the theory of HOG](https://www.learnopencv.com/histogram-of-oriented-gradients/)

```
import cv2
import dlib
from matplotlib import pyplot as plt

frame = cv2.imread("gdrive/My Drive/Colab Notebooks/face_detection/photos/648_oval_celebs.jpg")
detector = dlib.get_frontal_face_detector()

facesDetected = detector(frame, 3)

print(facesDetected)
print("Faces Detected: ", len(facesDetected))
for face in facesDetected:
    #print(face)
    #print(face.left())
    #print(face.top())
    #print(face.right())
    #print(face.bottom())
    left, top, right, botton = (int(face.left()), int(face.top()), int(face.right()), int(face.bottom()))
    #Drawing the bounding boxs:
    cv2.rectangle(frame, (left, top), (right, botton), (0, 255, 255), 2)

frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(frame)
plt.show()
```

This detector is also able to identify the direction to where the detected face is looking:
 ![alt text](http://1.bp.blogspot.com/-SJE_QOGqb_I/UNM8c5XbdhI/AAAAAAAAEj4/KqsLU9PVUgM/s1600/heads3.jpg)

```
import cv2
import dlib
from matplotlib import pyplot as plt

subdetector = ["looking ahead", "looking left", "looking right",
               "looking ahead and turning left", "looking ahead and turning right"]

frame = cv2.imread("gdrive/My Drive/Colab Notebooks/face_detection/photos/faceAngles_2.jpg")
detector = dlib.get_frontal_face_detector()

#image scale: 1 ... thresold detection -0.7
facesDetected, score, idx = detector.run(frame, 1, -0.7)
#print(facesDetected)
#print(score)
#print(idx)
for i, d in enumerate(facesDetected):
    #print(i)
    #print(d)
    print("Detection: {}, score: {}, Sub-detector: {}".format(d, score[i], subdetector[idx[i]]))
    left, top, right, botton = (int(d.left()), int(d.top()), int(d.right()), int(d.bottom()))
    cv2.rectangle(frame, (left, top), (right, botton), (0, 0, 255), 2)

frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
plt.imshow(frame)
plt.show()
```

# Anti-Surveillance Camouflage:

The patterns applied the paint is important: To the pixel-calculating machinations of facial recognition algorithms, they transformed the face into a mess of unremarkable pixels. In the computer’s vision, this camouflage caused a momentary burst of confusion.

[More about Anti-Surveillance Camouflage: https://www.theatlantic.com/technology/archive/2014/07/makeup/374929/](https://)

```
import cv2
import dlib
from matplotlib import pyplot as plt

frame1 = cv2.imread("gdrive/My Drive/Colab Notebooks/face_detection/photos/NOAnti-SurveillanceCamouflage.jpeg")
NOCamouflage = cv2.cvtColor(frame1, cv2.COLOR_BGR2RGB)
frame2 = cv2.imread("gdrive/My Drive/Colab Notebooks/face_detection/photos/Anti-SurveillanceCamouflage.jpg")
Camouflage  = cv2.cvtColor(frame2, cv2.COLOR_BGR2RGB)

detector = dlib.get_frontal_face_detector()

facesDetected = detector(NOCamouflage, 3)

print(facesDetected)
print("Faces Detected: ", len(facesDetected))
for face in facesDetected:
    #print(face)
    #print(face.left())
    #print(face.top())
    #print(face.right())
    #print(face.bottom())
    left, top, right, botton = (int(face.left()), int(face.top()), int(face.right()), int(face.bottom()))
    #Drawing the bounding boxs:
    cv2.rectangle(NOCamouflage, (left, top), (right, botton), (0, 255, 255), 2)

plt.imshow(NOCamouflage)
plt.show()

facesDetected = detector(Camouflage, 3)

print(facesDetected)
print("Faces Detected: ", len(facesDetected))
for face in facesDetected:
    #print(face)
    #print(face.left())
    #print(face.top())
    #print(face.right())
    #print(face.bottom())
    left, top, right, botton = (int(face.left()), int(face.top()), int(face.right()), int(face.bottom()))
    #Drawing the bounding boxs:
    cv2.rectangle(Camouflage, (left, top), (right, botton), (0, 255, 255), 2)

plt.imshow(Camouflage)
plt.show()
```
